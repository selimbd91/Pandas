```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
f = 50
T = 1 / f
x = np.linspace(0, 2*T, 100)
y = np.cos(2*np.pi*f*x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("sinx ar graph")
#fig.show()
```

```python
fig, ax = plt.subplots(figsize= (12, 5))
xx = 0
for i in range(1, 5):
    y = np.cos(2*np.pi*f*x *i) + np.sin(2*np.pi*f*x *i)
    xx += y
    ax.plot(x, y)
    
#plt.plot(x, xx)
```
