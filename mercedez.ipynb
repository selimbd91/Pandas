```python
def one_hot_encodding(df, variable, top_ten):
    for i in top_ten:
        df[var+'_'+i] = np.where(df[variable] == i, 1, 0)
        
df = pd.read_csv('train.csv', usecols= ['X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6'])
one_hot_encodding(df, 'X1', top_10_X1,)
one_hot_encodding(df, 'X2', top_10_X2)
df.drop(columns=df.columns[0:7])
```
